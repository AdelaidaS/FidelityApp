```python
#informe por feature
def data_report(infiel):
    # Sacamos los NOMBRES
    cols = pd.DataFrame(infiel.columns.values, columns=["COL_N"])

    # Sacamos los TIPOS
    types = pd.DataFrame(infiel.dtypes.values, columns=["DATA_TYPE"])

    # Sacamos los MISSINGS
    percent_missing = round(infiel.isnull().sum() * 100 / len(infiel), 2)
    percent_missing_infiel = pd.DataFrame(percent_missing.values, columns=["MISSINGS (%)"])

    # Sacamos los VALORES UNICOS
    unicos = pd.DataFrame(infiel.nunique().values, columns=["UNIQUE_VALUES"])
    
    percent_cardin = round(unicos['UNIQUE_VALUES']*100/len(infiel), 2)
    percent_cardin_infiel = pd.DataFrame(percent_cardin.values, columns=["CARDIN (%)"])

    concatenado = pd.concat([cols, types, percent_missing_infiel, unicos, percent_cardin_infiel], axis=1, sort=False)
    concatenado.set_index('COL_N', drop=True, inplace=True)


    return concatenado.T
        
data_report(infiel)
```

```python
cols=5
rows = int(np.ceil(float(infiel.shape[1]) / cols))
fig, axs = plt.subplots(nrows=rows, ncols=cols, figsize=(20, 10))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']

for i, column in enumerate(infiel.columns):
    row = i // cols
    col = i % cols
    ax = axs[row, col]
    ax.set_title(column)
    if infiel.dtypes[column] == np.object:
        infiel[column].value_counts().plot(kind="bar", ax=ax, color=colors[i%10])
    else:
        infiel[column].hist(ax=ax, color=colors[i%10])
        ax.tick_params(axis='x', rotation=45)

plt.subplots_adjust(hspace=0.7, wspace=0.2)
plt.show()
```
